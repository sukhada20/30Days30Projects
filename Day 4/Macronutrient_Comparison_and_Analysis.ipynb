```python
nutrient_columns = [
    'Vitamin C (mg per 100g)',
    'Vitamin B11 (mg per 100g)',
    'Sodium (mg per 100g)',
    'Calcium (mg per 100g)',
    'Carbohydrates (g per 100g)',
    'Iron (mg per 100g)',
    'Calories (kcal per 100g)',
    'Sugars (g per 100g)',
    'Dietary Fiber (g per 100g)',
    'Fat (g per 100g)',
    'Protein (g per 100g)'
]

df['nutrient_score'] = df[nutrient_columns].sum(axis=1)

top20_nutrient_dense = df.nlargest(20, 'nutrient_score')

top20_nutrient_dense[['food_normalized', 'nutrient_score']]
```

```python
sns.barplot(
    x='nutrient_score',
    y='food_normalized',
    data=top20_nutrient_dense,
    palette='magma'
)
plt.title("Top 20 Nutrient-Dense Foods")
plt.xlabel("Total Nutrient Score")
plt.ylabel("Food")
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
nutrient_cols = [
    'Vitamin C (mg per 100g)', 'Vitamin B11 (mg per 100g)', 'Sodium (mg per 100g)',
    'Calcium (mg per 100g)', 'Carbohydrates (g per 100g)', 'Iron (mg per 100g)',
    'Calories (kcal per 100g)', 'Sugars (g per 100g)', 'Dietary Fiber (g per 100g)',
    'Fat (g per 100g)', 'Protein (g per 100g)'
]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[nutrient_cols])
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
df['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=df['cluster'], palette='Set2')
plt.title("Food Clusters Based on Nutritional Content (PCA 2D)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title="Cluster")
plt.tight_layout()
plt.show()
```

```python
df['cluster'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title("Number of Foods in Each Cluster")
plt.xlabel("Cluster")
plt.ylabel("Number of Foods")
plt.tight_layout()
plt.show()
```
